```python
from NanTex_backend.data_preparation import img_augmentation as img_augmentation
from NanTex_backend.batching.BatchDataLoader import BatchDataLoader_Handler
```

```python
## Setup image augmentation
train_augmentation_pipeline = img_augmentation.train_augmentation_pipeline
val_augmentation_pipeline = img_augmentation.val_augmentation_pipeline
```

```python
## Generate boilerplate config file
BatchDataLoader_Handler.generate_biolerplate_config_file()
```

```python
## Setup BatchDataLoader
BDL: BatchDataLoader_Handler
BDL = BatchDataLoader_Handler.from_config(
    config_file_path="./configs/BatchDataLoader_wurzburg_autobatch_config.json",
    DEBUG=True,
)

BDL.setup_augmentation_pipelines(
    raw_aug=train_augmentation_pipeline, val_aug=val_augmentation_pipeline
)

train_gen, val_gen = BDL.build_BatchDataLoader()
```

```python
## Check the data
# no-autobatch

import matplotlib.pyplot as plt
import numpy as np

# no-autobatch
plt.style.use("dark_background")
x, y = train_gen.dataset[0]
x, y = np.asanyarray(x), np.asanyarray(y)
print("X", np.max(x), np.min(x))
print("X_mean", np.mean(x), "X_std", np.std(x))
print("y1", np.max(y[0, 0, ...]), np.min(y[0, 0, ...]))
print("y1_mean", np.mean(y[0, 0, ...]), "y1_std", np.std(y[0, 0, ...]))
print("y2", np.max(y[1, 0, ...]), np.min(y[1, 0, ...]))
print("y2_mean", np.mean(y[1, 0, ...]), "y2_std", np.std(y[1, 0, ...]))
print("y3", np.max(y[2, 0, ...]), np.min(y[2, 0, ...]))
print("y3_mean", np.mean(y[2, 0, ...]), "y3_std", np.std(y[2, 0, ...]))
fig, ax = plt.subplots(1, 4, figsize=(10, 5), dpi=200)
ax[0].imshow(x[0, 0, ...])
ax[0].axis("off")

ax[1].imshow(y[0, 0])
ax[1].axis("off")

ax[2].imshow(y[0, 1])
ax[2].axis("off")

ax[3].imshow(y[0, 2])
ax[3].axis("off")
```

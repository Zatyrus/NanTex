```python
@ray.remote
def imagine(path):
    Img = np.load(path)
    plt.style.use('dark_background')
    plt.subplots_adjust(wspace=0.3, top=1)
    font = {'family' : 'Courier new',
            'weight' : 'bold',
            'size'   : 10}
    colormap = 'gray'

    matplotlib.rc('font', **font)

    fig, ((ax1, ax2,ax3,ax7),(ax4, ax5,ax6,ax8)) = plt.subplots(2,4,figsize = (12,5), dpi = 500)

    ax1.axis('off')
    ax1.add_artist(ScaleBar(10, units='nm', location = 'lower right', frameon=0, color = 'w', pad = 0 ,font_properties = {'family' : 'Courier new','weight' : 'bold','size'   : 5} ))
    ax1.set_title('Overlay GT')
    gt = Img[:,:,0]+Img[:,:,1]+Img[:,:,2]
    gt = gt/np.max(gt) * 255
    im1 = ax1.imshow(gt, cmap = colormap)
    plt.colorbar(im1, ax = ax1)

    ax2.axis('off')
    ax2.add_artist(ScaleBar(10, units='nm', location = 'lower right', frameon=0, color = 'w', pad = 0 ,font_properties = {'family' : 'Courier new','weight' : 'bold','size'   : 5} ))
    ax2.set_title('Ground truth ER')
    im2 = ax2.imshow(Img[:,:,0], cmap = colormap)
    plt.colorbar(im2, ax = ax2)

    ax3.axis('off')
    ax3.add_artist(ScaleBar(10, units='nm', location = 'lower right', frameon=0, color = 'w', pad = 0 ,font_properties = {'family' : 'Courier new','weight' : 'bold','size'   : 5} ))
    ax3.set_title('Ground truth MIC')
    im3 = ax3.imshow(Img[:,:,1], cmap = colormap)
    plt.colorbar(im3, ax = ax3)

    ax7.axis('off')
    ax7.add_artist(ScaleBar(10, units='nm', location = 'lower right', frameon=0, color = 'w', pad = 0 ,font_properties = {'family' : 'Courier new','weight' : 'bold','size'   : 5} ))
    ax7.set_title('Ground truth CLA')
    im7 = ax7.imshow(Img[:,:,2], cmap = colormap)
    plt.colorbar(im7, ax = ax7)

    ax4.axis('off')
    ax4.add_artist(ScaleBar(10, units='nm', location = 'lower right', frameon=0, color = 'w', pad = 0 ,font_properties = {'family' : 'Courier new','weight' : 'bold','size'   : 5} ))
    ax4.set_title('Overlay Segmented')
    seg = Img[:,:,3]+Img[:,:,4]+Img[:,:,5]
    seg = seg/np.max(seg)*255
    im4 = ax4.imshow(seg, cmap = colormap)
    plt.colorbar(im4, ax = ax4)

    ax5.axis('off')
    ax5.add_artist(ScaleBar(10, units='nm', location = 'lower right', frameon=0, color = 'w', pad = 0 ,font_properties = {'family' : 'Courier new','weight' : 'bold','size'   : 5} ))
    ax5.set_title('Segmented ER')
    im5 = ax5.imshow(Img[:,:,3], cmap = colormap)
    plt.colorbar(im5, ax = ax5)

    ax6.axis('off')
    ax6.add_artist(ScaleBar(10, units='nm', location = 'lower right', frameon=0, color = 'w', pad = 0 ,font_properties = {'family' : 'Courier new','weight' : 'bold','size'   : 5} ))
    ax6.set_title('Segmented MIC')
    im6 = ax6.imshow(Img[:,:,4], cmap = colormap)
    plt.colorbar(im6, ax = ax6)

    ax8.axis('off')
    ax8.add_artist(ScaleBar(10, units='nm', location = 'lower right', frameon=0, color = 'w', pad = 0 ,font_properties = {'family' : 'Courier new','weight' : 'bold','size'   : 5} ))
    ax8.set_title('Segmented CLA')
    im8 = ax8.imshow(Img[:,:,5], cmap = colormap)
    plt.colorbar(im8, ax = ax8)

    plt.savefig(f'../../../Paper Prep/Wuerzburg/png/{os.path.basename(path)[:-4]}.png')
```

```python
from skimage.io import imread
def imagine_single(path):
    Img = np.load(path)
    plt.style.use('dark_background')
    font = {'family' : 'Courier new',
            'weight' : 'bold',
            'size'   : 13}
    colormap = 'gray'
    
    tag = 'Würzburg_OVL_MIC_ER'

    matplotlib.rc('font', **font)
    
    for i,label in enumerate(['Segmented ER', 'Segmented MIC', 'Segmented CLA']): #['Overlay GT', 'skip', 'Ground truth MIC', 'skip', 'skip', 'Ground truth LNPs', 'OVL MIC-Clustering', 'OVL LNP-Clustering', 'MIC MIC-Clustering', 'MIC LNP-Clustering', 'LNPs MIC-Clustering', 'LNPs LNP-Clustering']
        fig, ax = plt.subplots(1,1,figsize=(8,8), dpi = 300)
        ax.axis('off')
        #ax.set_title(label)
        ax.imshow(Img[...,i], cmap=colormap)
        #ax.add_artist(ScaleBar(10, units='nm', location = 'lower right', frameon=0, color = 'w', pad = 0 ,font_properties = {'family' : 'Courier new','weight' : 'bold','size'   : 9} ))
        plt.savefig(f"{tag}_{label}.png", bbox_inches = 'tight', transparent=True)
        plt.close()
    
    fig, ax = plt.subplots(1,1,figsize = (8,8), dpi = 300)

    # ax.axis('off')
    # #ax.add_artist(ScaleBar(10, units='nm', location = 'lower right', frameon=0, color = 'w', pad = 0 ,font_properties = {'family' : 'Courier new','weight' : 'bold','size'   : 9} ))
    # #ax.set_title('Overlay Seg')
    # ovl = Img[:,:,3]+Img[:,:,4]+Img[:,:,5]
    # ovl = ovl/np.max(ovl) * 255
    # ax.imshow(ovl, cmap = colormap)
    # plt.savefig(f'{tag}_Overlay_seg.png', bbox_inches = 'tight', transparent=True)
    # plt.close()
    
    # fig, ax = plt.subplots(1,1,figsize = (8,8), dpi = 300)

    ax.axis('off')
    #ax.add_artist(ScaleBar(10, units='nm', location = 'lower right', frameon=0, color = 'w', pad = 0 ,font_properties = {'family' : 'Courier new','weight' : 'bold','size'   : 9} ))
    #ax.set_title('Overlay GT')
    gt = Img[:,:,0]+Img[:,:,1]+Img[:,:,2]
    #gt = gt/np.max(gt) * 255
    ax.imshow(gt, cmap = colormap)
    plt.savefig(f'{tag}_Overlay_Seg.png', bbox_inches = 'tight', transparent=True)
    plt.close()
    
    fig, ax = plt.subplots(1,1,figsize = (8,8), dpi = 300)

    ax.axis('off')
    #ax.add_artist(ScaleBar(10, units='nm', location = 'lower right', frameon=0, color = 'w', pad = 0 ,font_properties = {'family' : 'Courier new','weight' : 'bold','size'   : 9} ))
    #ax.set_title('Overlay GT')
    Img[:,:,0] = Img[:,:,0]/np.max(Img[:,:,0]) * 255
    Img[:,:,1] = Img[:,:,1]/np.max(Img[:,:,1]) * 255
    Img[:,:,2] = Img[:,:,2]/np.max(Img[:,:,2]) * 255
    gt = Img[:,:,0]+Img[:,:,1]+Img[:,:,2]
    #gt = gt/np.max(gt) * 255
    ax.imshow(gt, cmap = colormap)
    plt.savefig(f'{tag}_Overlay_GT_train.png', bbox_inches = 'tight', transparent=True)
    plt.close()
```

```python
import pickle
num_features = 3
average = {}

for i in range(num_features):
    average[f'feature_{i}'] = {
        'PSNR':[],
        'SSIM':[], 
        'SSIM_sewar':[], 
        'MSSSIM':[], 
        'FSIM':[]}

for file in files:
    with open(file, 'rb') as current:
        current = pickle.load(current)
        for key in current.keys():
            for sub_key in current[key].keys():
                average[key][sub_key].append(current[key][sub_key])

for key in average.keys():
    for sub_key in average[key].keys():
        average[key][sub_key] = np.average(average[key][sub_key])
```

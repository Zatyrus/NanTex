```python
import os
def normalize(Source, Target):
    m = []
    v = []
        
    for file in Source:
        current = np.load(file)

        if len(current.shape) == 3:
            current = current[...,0]
        m.append(np.mean(current, axis=(0,1)))
        v.append(np.std(current, axis=(0,1)))
        
    print(m,v)
                
    m = np.average(np.array(m))
    v = np.average(np.array(v))

    
    for file in Source:
        current = np.load(file)

        if len(current.shape) == 3:
            current = current[...,0]
        
        current = (current-m)/v
    
        np.save(file =  f'{Target}/normalized_{os.path.basename(file)}', arr=current)
```

```python
from tqdm.notebook import tqdm
def overlay(M_SOURCE,L_SOURCE,A_SOURCE, mode:str = 'raw_mix'):
    for i,MICRO in enumerate(tqdm(M_SOURCE)):
        for j,LNP in enumerate(tqdm(L_SOURCE)):
            for t,ACTIN in enumerate(tqdm(A_SOURCE)):
                MICRO_IMG = np.load(MICRO)
                LNP_IMG = np.load(LNP)
                ACTIN_IMG = np.load(ACTIN)#[:,:,0]
                
                y_lim = min(MICRO_IMG.shape[0],LNP_IMG.shape[0],ACTIN_IMG.shape[0]) 
                x_lim = min(MICRO_IMG.shape[1],LNP_IMG.shape[1],ACTIN_IMG.shape[1]) 
                
                MICRO_IMG = MICRO_IMG[:y_lim,:x_lim]
                LNP_IMG = LNP_IMG[:y_lim,:x_lim]
                ACTIN_IMG = ACTIN_IMG[:y_lim,:x_lim]
                
                out = np.zeros(shape=(4,y_lim,x_lim))
                
                for p,img in enumerate([MICRO_IMG, LNP_IMG, ACTIN_IMG]):
                    out[p+1,:,:] = img
                    
                out[0,:,:] = MICRO_IMG + LNP_IMG + ACTIN_IMG
                
                for k in range(0,4):
                    mix = MICRO_IMG + np.rot90(LNP_IMG,k = k)
                    out[2,:,:] = np.rot90(LNP_IMG, k = k)
                    
                    for s in range(0,4):
                        out[0,:,:] = mix + np.rot90(ACTIN_IMG, k = s)
                        out[3,:,:] = np.rot90(ACTIN_IMG, k = s)
                        np.save(f"./{mode}/ImMIX_ER-MIC-CLT_({i},{j},{t}_({k},{s})).npy", out.astype('float32'))
```

```python
from tqdm.notebook import tqdm

def pad_img(img, y_lim, x_lim, limit_to):
    y_pad = max(0,int((y_lim-img.shape[0])/2))
    x_pad = max(0,int((x_lim-img.shape[1])/2))
    out = np.zeros(shape=(y_lim,x_lim))
    out[y_pad:img.shape[0]+y_pad, x_pad:img.shape[1]+x_pad] = img
    return out[int((y_lim-limit_to)//2):-(int((y_lim-limit_to)//2)+1),int((x_lim-limit_to)//2):-(int((x_lim-limit_to)//2)+1)]


def overlay_with_pad(M_SOURCE,L_SOURCE,A_SOURCE, mode:str = 'raw_mix',limit_to=4000):
    for i,MICRO in enumerate(tqdm(M_SOURCE[:3])):
        for j,LNP in enumerate(tqdm(L_SOURCE[:3])):
            for t,ACTIN in enumerate(tqdm(A_SOURCE[:3])):
                MICRO_IMG = np.load(MICRO)
                LNP_IMG = np.load(LNP)
                ACTIN_IMG = np.load(ACTIN)#[:,:,0]
                
                lim = max(MICRO_IMG.shape[0],LNP_IMG.shape[0],ACTIN_IMG.shape[0],MICRO_IMG.shape[1],LNP_IMG.shape[1],ACTIN_IMG.shape[1]) 
                
                MICRO_IMG = pad_img(MICRO_IMG,lim,lim,limit_to)
                LNP_IMG = pad_img(LNP_IMG,lim,lim,limit_to)
                ACTIN_IMG = pad_img(ACTIN_IMG,lim,lim,limit_to)
                
                #print(MICRO_IMG.shape, LNP_IMG.shape, ACTIN_IMG.shape)
                
                out = np.zeros(shape=(4,limit_to,limit_to)) #(4,lim,lim)
                
                for p,img in enumerate([MICRO_IMG, LNP_IMG, ACTIN_IMG]):
                    out[p+1,:,:] = img
                    
                out[0,:,:] = MICRO_IMG + LNP_IMG + ACTIN_IMG
                
                for k in range(0,4):
                    mix = MICRO_IMG + np.rot90(LNP_IMG,k = k)
                    out[2,:,:] = np.rot90(LNP_IMG, k = k)
                    
                    for s in range(0,4):
                        out[0,:,:] = mix + np.rot90(ACTIN_IMG, k = s)
                        out[3,:,:] = np.rot90(ACTIN_IMG, k = s)
                        np.save(f"./{mode}/ImMIX_ER-MIC-CLT_({i},{j},{t}_({k},{s})).npy", out.astype('float32'))
                        
def overlay_with_pad_no_tilt(M_SOURCE,L_SOURCE,A_SOURCE, mode:str = 'raw_mix'):
    for i,MICRO in enumerate(tqdm(M_SOURCE)):
        for j,LNP in enumerate(tqdm(L_SOURCE)):
            for t,ACTIN in enumerate(tqdm(A_SOURCE)):
                MICRO_IMG = np.load(MICRO)
                LNP_IMG = np.load(LNP)
                ACTIN_IMG = np.load(ACTIN)#[:,:,0]
                
                lim = max(MICRO_IMG.shape[0],LNP_IMG.shape[0],ACTIN_IMG.shape[0],MICRO_IMG.shape[1],LNP_IMG.shape[1],ACTIN_IMG.shape[1]) 
                
                MICRO_IMG = pad_img(MICRO_IMG,lim,lim)
                LNP_IMG = pad_img(LNP_IMG,lim,lim)
                ACTIN_IMG = pad_img(ACTIN_IMG,lim,lim)
                
                #print(MICRO_IMG.shape, LNP_IMG.shape, ACTIN_IMG.shape)
                
                out = np.zeros(shape=(4,lim,lim))
                
                for p,img in enumerate([MICRO_IMG, LNP_IMG, ACTIN_IMG]):
                    out[p+1,:,:] = img
                    
                out[0,:,:] = MICRO_IMG + LNP_IMG + ACTIN_IMG
                
                np.save(f"{mode}/ImMIX_ER-MIC-CLT_({i},{j},{t}).npy", out.astype('float32'))
                
import ray

@ray.remote
def overlay_with_pad_ray(M_SOURCE,L_SOURCE,A_SOURCE, i,j,t, mode:str = 'raw_mix',limit_to=4000):
    MICRO_IMG = np.load(M_SOURCE)[:2601,:2601]
    LNP_IMG = np.load(L_SOURCE)[:2601,:2601]
    ACTIN_IMG = np.load(A_SOURCE)[:2601,:2601]
    
    lim = max(MICRO_IMG.shape[0],LNP_IMG.shape[0],ACTIN_IMG.shape[0],MICRO_IMG.shape[1],LNP_IMG.shape[1],ACTIN_IMG.shape[1]) 
    
    MICRO_IMG = pad_img(MICRO_IMG,lim,lim,limit_to)
    LNP_IMG = pad_img(LNP_IMG,lim,lim,limit_to)
    ACTIN_IMG = pad_img(ACTIN_IMG,lim,lim,limit_to)
    
    ## new single image normalization
    MICRO_IMG = MICRO_IMG/np.max(MICRO_IMG) * 255
    LNP_IMG = LNP_IMG/np.max(LNP_IMG) * 255
    ACTIN_IMG = ACTIN_IMG/np.max(ACTIN_IMG) * 255
        
    out = np.zeros(shape=(4,limit_to,limit_to))
        
    out[1,:,:] = MICRO_IMG
    
    for k in range(0,4):
        mix = MICRO_IMG + np.rot90(LNP_IMG,k = k)
        out[2,:,:] = np.rot90(LNP_IMG, k = k)
        
        for s in range(0,4):
            out[0,:,:] = mix + np.rot90(ACTIN_IMG, k = s)
            out[3,:,:] = np.rot90(ACTIN_IMG, k = s)
            np.save(f"{mode}/ImMIX_POST_ER-MIC-CLT_({i},{j},{t}_({k},{s})_max({np.max(out[0,...])})).npy", out.astype('uint16'))
```

```python
def model_step(model, loss_fn, optimizer, feature, label, activation, prediction_type=None, train_step=True):
    
    # zero gradients if training
    if train_step:
        optimizer.zero_grad()
    
    # forward
    logits = model(feature)
        
    # final activation
    predicted = activation(logits)

    # pass through loss
    loss_value = loss_fn(input=predicted, target=label)
 
    # backward if training mode
    if train_step:
        loss_value.backward()
        optimizer.step()
    
    return loss_value, predicted
```

```python
import torch
import torch.nn.functional as F
from math import exp
import numpy as np


def gaussian(window_size, sigma):
    gauss = torch.Tensor([exp(-(x - window_size//2)**2/float(2*sigma**2)) for x in range(window_size)])
    return gauss/gauss.sum()


def create_window(window_size, channel=1):
    _1D_window = gaussian(window_size, 1.5).unsqueeze(1)
    _2D_window = _1D_window.mm(_1D_window.t()).float().unsqueeze(0).unsqueeze(0)
    window = _2D_window.expand(channel, 1, window_size, window_size).contiguous()
    return window


def ssim(img1, img2, window_size=11, window=None, size_average=True, full=False, val_range=None):
    # Value range can be different from 255. Other common ranges are 1 (sigmoid) and 2 (tanh).
    if val_range is None:
        if torch.max(img1) > 128:
            max_val = 255
        else:
            max_val = 1

        if torch.min(img1) < -0.5:
            min_val = -1
        else:
            min_val = 0
        L = max_val - min_val
    else:
        L = val_range

    padd = 0
    (_, channel, height, width) = img1.size()
    if window is None:
        real_size = min(window_size, height, width)
        window = create_window(real_size, channel=channel).to(img1.device)

    mu1 = F.conv2d(img1, window, padding=padd, groups=channel)
    mu2 = F.conv2d(img2, window, padding=padd, groups=channel)

    mu1_sq = mu1.pow(2)
    mu2_sq = mu2.pow(2)
    mu1_mu2 = mu1 * mu2

    sigma1_sq = F.conv2d(img1 * img1, window, padding=padd, groups=channel) - mu1_sq
    sigma2_sq = F.conv2d(img2 * img2, window, padding=padd, groups=channel) - mu2_sq
    sigma12 = F.conv2d(img1 * img2, window, padding=padd, groups=channel) - mu1_mu2

    C1 = (0.01 * L) ** 2
    C2 = (0.03 * L) ** 2

    v1 = 2.0 * sigma12 + C2
    v2 = sigma1_sq + sigma2_sq + C2
    cs = torch.mean(v1 / v2)  # contrast sensitivity

    ssim_map = ((2 * mu1_mu2 + C1) * v1) / ((mu1_sq + mu2_sq + C1) * v2)

    if size_average:
        ret = ssim_map.mean()
    else:
        ret = ssim_map.mean(1).mean(1).mean(1)

    if full:
        return ret, cs
    return ret


def msssim(img1, img2, window_size=11, size_average=True, val_range=None, normalize=False):
    device = img1.device
    weights = torch.FloatTensor([0.0448, 0.2856, 0.3001, 0.2363, 0.1333]).to(device)
    # weights = torch.FloatTensor([0.0448, 0.2856, 0.3001, 0.2363, 0.1333])
    levels = weights.size()[0]
    mssim = []
    mcs = []
    for _ in range(levels):
        sim, cs = ssim(img1, img2, window_size=window_size, size_average=size_average, full=True, val_range=val_range)
        #print("sim",sim)
        mssim.append(sim)
        mcs.append(cs)

        img1 = F.avg_pool2d(img1, (2, 2))
        img2 = F.avg_pool2d(img2, (2, 2))

    mssim = torch.stack(mssim)
    mcs = torch.stack(mcs)

    # Normalize (to avoid NaNs during training unstable models, not compliant with original definition)
    if normalize:
        mssim = (mssim + 1) / 2
        mcs = (mcs + 1) / 2

    pow1 = mcs ** weights
    pow2 = mssim ** weights
    # From Matlab implementation https://ece.uwaterloo.ca/~z70wang/research/iwssim/
    output = torch.prod(pow1[:-1] * pow2[-1])
    return output


# Classes to re-use window
class SSIM(torch.nn.Module):
    def __init__(self, window_size=11, size_average=True, val_range=None):
        super(SSIM, self).__init__()
        self.window_size = window_size
        self.size_average = size_average
        self.val_range = val_range

        # Assume 1 channel for SSIM
        self.channel = 1
        self.window = create_window(window_size)

    def forward(self, img1, img2):
        (_, channel, _, _) = img1.size()

        if channel == self.channel and self.window.dtype == img1.dtype:
            window = self.window
        else:
            window = create_window(self.window_size, channel).to(img1.device).type(img1.dtype)
            self.window = window
            self.channel = channel

        return ssim(img1, img2, window=window, window_size=self.window_size, size_average=self.size_average)

class MSSSIM(torch.nn.Module):
    def __init__(self, window_size=11, size_average=True, channel=3):
        super(MSSSIM, self).__init__()
        self.window_size = window_size
        self.size_average = size_average
        self.channel = channel

    def forward(self, img1, img2):
        # TODO: store window between calls if possible,
        # return msssim(img1, img2, window_size=self.window_size, size_average=self.size_average)
        return msssim(img1, img2, window_size=self.window_size, size_average=self.size_average, normalize=True)
```

```python
def model_step_redundance(model, loss_fn, optimizer, feature, label, activation, L1:float = None, prediction_type=None, train_step=True):
    
    # zero gradients if training
    if train_step:
        optimizer.zero_grad()
    
    # forward
    predicted = model(feature)

    # pass through loss
    loss_value = loss_fn(input=predicted, target=label)
 
    # backward if training mode
    if train_step:
        loss_value.backward()
        optimizer.step()
        
    # Additional redundance step
    predicted = model(predicted)

    # pass through loss
    loss_value = loss_fn(input=predicted, target=label)
 
    # backward if training mode
    if train_step:
        loss_value.backward()
        optimizer.step()

    outputs = {
        'pred': predicted
    }
    
    return loss_value, outputs
```

```python
# Utility function to evaluate our segmentations

from scipy.optimize import linear_sum_assignment
from skimage.segmentation import relabel_sequential

def evaluate(gt_labels, pred_labels, th=0.5):
    pred_labels_rel, _, _ = relabel_sequential(pred_labels)
    gt_labels_rel, _, _ = relabel_sequential(gt_labels)

    overlay = np.array([pred_labels_rel.flatten(),
                        gt_labels_rel.flatten()])

    # get overlaying cells and the size of the overlap
    overlay_labels, overlay_labels_counts = np.unique(
        overlay, return_counts=True, axis=1)
    overlay_labels = np.transpose(overlay_labels)

    # get gt cell ids and the size of the corresponding cell
    gt_labels_list, gt_counts = np.unique(gt_labels_rel, return_counts=True)
    gt_labels_count_dict = {}
    
    for (l, c) in zip(gt_labels_list, gt_counts):
        gt_labels_count_dict[l] = c

    # get pred cell ids
    pred_labels_list, pred_counts = np.unique(pred_labels_rel,
                                              return_counts=True)

    pred_labels_count_dict = {}
    for (l, c) in zip(pred_labels_list, pred_counts):
        pred_labels_count_dict[l] = c

    num_pred_labels = int(np.max(pred_labels_rel))
    num_gt_labels = int(np.max(gt_labels_rel))
    num_matches = min(num_gt_labels, num_pred_labels)
    
    # create iou table
    iouMat = np.zeros((num_gt_labels+1, num_pred_labels+1),
                      dtype=np.float32)

    for (u, v), c in zip(overlay_labels, overlay_labels_counts):
        iou = c / (gt_labels_count_dict[v] + pred_labels_count_dict[u] - c)
        iouMat[int(v), int(u)] = iou

    # remove background
    iouMat = iouMat[1:, 1:]

    # use IoU threshold th
    if num_matches > 0 and np.max(iouMat) > th:
        costs = -(iouMat > th).astype(float) - iouMat / (2*num_matches)
        gt_ind, pred_ind = linear_sum_assignment(costs)
        assert num_matches == len(gt_ind) == len(pred_ind)
        match_ok = iouMat[gt_ind, pred_ind] > th
        tp = np.count_nonzero(match_ok)
    else:
        tp = 0
    fp = num_pred_labels - tp
    fn = num_gt_labels - tp
    ap = tp / max(1, tp + fn + fp)
    precision = tp / max(1, tp + fp)
    recall = tp / max(1, tp + fn)

    return ap, precision, recall, tp, fp, fn
```

```python
def train(train_loader, val_loader, net, loss_fn, activation, optimizer, dtype, writer, device, training_steps, log_interval:int = 100, prediction_type=None, save_dir:str = './model'):

    # set train flags, initialize step
    net.train() 
    loss_fn.train()
    step = 0
    print_loss = 0
    best_loss = np.infty
    best_msssim = 0
    
    SSIM_Metr = SSIM()
    SSIM_Metr.to(device)
    MSSSIM_Metr = MSSSIM()
    MSSSIM_Metr.to(device)


    with tqdm(total=training_steps, leave=False, desc='Current batch loss: 0') as pbar:
        while step < training_steps:
            # reset data loader to get random augmentations
            np.random.seed()
            tmp_loader = iter(train_loader)
            for feature, label in tmp_loader:
                label = label.type(dtype)
                label = label.to(device)
                feature = feature.type(dtype)
                feature = feature.to(device)
                loss_value, pred = model_step(model=net, loss_fn=loss_fn, optimizer=optimizer, feature=feature, label=label, activation=activation, prediction_type = None, train_step=True)
                writer.add_scalar('MSE',loss_value.cpu().detach().numpy(),step)
                writer.add_scalar('SSIM',SSIM_Metr(pred,label).cpu().detach().numpy(),step)
                writer.add_scalar('MSSSIM',MSSSIM_Metr(pred,label).cpu().detach().numpy(),step)
                        
                step += 1
                pbar.set_description("Current batch loss: {:.3e}".format(loss_value.cpu().detach().numpy())) 
                pbar.update(1)
                
                if step % save_interval == 0:
                    torch.save(net.state_dict(), '../../../Paper Prep/models/paperPrep_WU_2023_lowerLR/save_{}.pt'.format(step))
                
                
                if step % log_interval == 0:
                    net.eval()
                    tmp_val_loader = iter(val_loader)
                    acc_loss = []
                    acc_ssim = []
                    acc_msssim = []
                    condition = 0
                    for feature, label in tmp_val_loader:   
                        pbar.set_description("Currently validating: {}".format(condition))                                          
                        label = label.type(dtype)
                        label = label.to(device)
                        feature = feature.type(dtype)
                        feature = feature.to(device)
                        loss_value, val_pred = model_step(model=net, loss_fn=loss_fn, optimizer=optimizer, feature=feature, label=label, activation=activation, prediction_type = None, train_step=False)
                        acc_loss.append(loss_value.cpu().detach().numpy())
                        acc_ssim.append(SSIM_Metr(val_pred,label).cpu().detach().numpy())
                        acc_msssim.append(MSSSIM_Metr(val_pred,label).cpu().detach().numpy())
                        
                        # if condition == 0:
                        #     l = validate(label.cpu().detach().numpy(), val_pred.cpu().detach().numpy(), l = None)
                        # else:
                        #     l = validate(label.cpu().detach().numpy(), val_pred.cpu().detach().numpy(), l = l)
                        
                        condition += 1
                        # if condition == 5:
                        #     break
                
                    writer.add_scalar(tag="val_MSE", scalar_value=np.mean(acc_loss), global_step=step)
                    writer.add_scalar('val_SSIM',scalar_value=np.mean(acc_ssim), global_step=step)
                    writer.add_scalar('val_MSSSIM',scalar_value=np.mean(acc_msssim), global_step=step)

                    #write(val_list=l, desc='val', writer=writer, step=step)

                                           
                    if np.mean(acc_loss) < best_loss:
                        best_loss = np.mean(acc_loss)
                        torch.save(net.state_dict(), f'{save_dir}/model_best.pt')
                        
                    if np.mean(acc_msssim) > best_msssim:
                        best_msssim = np.mean(acc_msssim)
                        torch.save(net.state_dict(), f'{save_dir}/model_best_msssim.pt')
                    
                    torch.save(net.state_dict(), f'{save_dir}/model_current.pt')
                    
                    net.train()
```
